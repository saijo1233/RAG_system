```python
import os
import time
import subprocess
import numpy as np
import faiss
import ollama
import gradio as gr
from pypdf import PdfReader

PDF_PATH = "/content/drive/MyDrive/RAG_System/5G_3GPP.pdf"
DB_PATH = "/content/drive/MyDrive/RAG_System/my_5g_database"
MODEL = "qwen3:4b"
EMBED = "bge-m3:latest"

!pkill ollama
os.environ["LD_LIBRARY_PATH"] = "/usr/lib64-nvidia"
subprocess.Popen(["ollama", "serve"], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
time.sleep(10)

client = ollama.Client(host='http://127.0.0.1:11434')

def load_data():
    idx_file = os.path.join(DB_PATH, "index.faiss")
    txt_file = os.path.join(DB_PATH, "texts.npy")

    if os.path.exists(idx_file):
        print("База найдена.")
        return faiss.read_index(idx_file), np.load(txt_file, allow_pickle=True).tolist()

    print("База данных не найдена.")
    reader = PdfReader(PDF_PATH)
    chunks = []
    for page in reader.pages:
        t = page.extract_text()
        if t:
            t = " ".join(t.split())
            for i in range(0, len(t), 700):
                chunks.append(t[i:i+800].strip())

    embs = []
    for i, c in enumerate(chunks):
        res = client.embeddings(model=EMBED, prompt=c)
        embs.append(res['embedding'])
        if i % 100 == 0: print(f"Готово: {i}/{len(chunks)}")

    if not os.path.exists(DB_PATH): os.makedirs(DB_PATH)
    d = len(embs[0])
    index = faiss.IndexFlatL2(d)
    index.add(np.array(embs).astype('float32'))

    faiss.write_index(index, idx_file)
    np.save(txt_file, np.array(chunks))
    return index, chunks

index, chunks = load_data()

def chat(msg, history):
    req = client.embeddings(model=EMBED, prompt=msg)
    vec = np.array([req['embedding']]).astype('float32')
    _, ids = index.search(vec, k=4)

    context = ""
    for i in ids[0]:
        context += chunks[i] + "\n---\n"

    prompt = (
        f"<|im_start|>system\n"
        f"Ты — ведущий инженер по 5G. Твоя задача: объяснить тему СВОИМИ СЛОВАМИ. "
        f"Используй предоставленные данные из 3GPP только как техническую базу. "
        f"ЗАПРЕЩЕНО копировать текст из документации целыми предложениями. "
        f"Твой ответ должен быть связным, аналитическим и профессиональным. Пиши на русском языке.<|im_end|>\n"
        f"<|im_start|>user\n"
        f"ТЕХНИЧЕСКИЕ ДАННЫЕ ДЛЯ АНАЛИЗА:\n{context}\n\n"
        f"ВОПРОС: {msg}<|im_end|>\n"
        f"<|im_start|>assistant\n"
    )

    ans = ""
    stream = client.generate(
        model=MODEL,
        prompt=prompt,
        stream=True,
        options={
            "temperature": 0.8,
            "repeat_penalty": 1.2,
            "stop": ["<|im_start|>", "<|im_end|>", "user\n"]
        }
    )

    for chunk in stream:
        ans += chunk['response']
        yield ans.split("</think>")[-1].lstrip()

demo = gr.ChatInterface(
    fn=chat,
    title="5G RAG System",
    description="Спрашивайте что угодно про 5G, я умный помощник.",
    examples=["Что такое 5G-AN?", "Для чего используется 5G-GUTI?", "Опиши архитектуру 5GS"]
)

if __name__ == "__main__":
    demo.launch(share=True)
```
